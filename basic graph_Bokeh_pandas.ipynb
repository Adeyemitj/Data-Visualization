```python
#Plotting a basic Bokeh line graph

from bokeh.plotting import figure
from bokeh.io import output_file, show
import pandas

df=pandas.read_csv("data.csv")
x=df["x"]
y=df["y"]

#prepare the output file
output_file("Line_from_csv.html")

#create a figure object
f=figure()

#create line plot
f.line(x,y)
f.triangle(x,y)
f.circle(x,y)

show(f)
```

```python
#Plotting percentage of women who received an engineering degree over years

import pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show

#prepare some data
df=pandas.read_csv("bachelors.csv")
x=df["Year"]
y=df["Engineering"]

#prepare the output file
output_file("bachhelors_degree.html")

#create a figure object
f=figure()

#create line, circle plot
f.line(x,y)
f.circle(x,y)

#write the plot in the figure object
show(f)

```

```python
#Plotting graph with title,width and height
import pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show

x=[1,2,3,12,26]
y=[4,5,6,23,30]

#create output file
output_file("mygraph.html")

#create figure object
p=figure(plot_width=500, plot_height=400, tools='pan')

p.title.text="Cool Data"
p.title.text_color="Green"
p.title.text_font="times"
p.title.text_font_style="bold"
p.xaxis.minor_tick_line_color="blue"
p.yaxis.minor_tick_line_color="blue"
p.xaxis.axis_label="Date"
p.yaxis.axis_label="Intensity"

p.line(x,y)
p.circle(x,y)

show(p)

```

```python
import pandas
from bokeh.plotting import figure, output_file, show
#from bokeh.io import output_file, show

df=pandas.read_excel("weather_data.xlsx")
df["Temperature"]=df["Temperature"]/10
df["Pressure"]=df["Pressure"]/10

output_file("Weather.html")

w=figure(plot_width=500, plot_height=400, tools='pan')

w.title.text="Temperature and Air Pressure"
w.title.text_color="Green"
w.title.text_font="Arial Black"
w.title.text_font_style="bold"
w.xaxis.minor_tick_line_color="Blue"
w.yaxis.minor_tick_line_color="Blue"
w.xaxis.axis_label="Temperature(oC)"
w.yaxis.axis_label="Pressure(hPa)"
w.xaxis.axis_label_text_color="Green"
w.yaxis.axis_label_text_color="Green"
w.xaxis.axis_label_text_font="Calibri"
w.yaxis.axis_label_text_font="Calibri"

w.triangle(df["Temperature"],df["Pressure"],size=0.5)
#w.circle(df["Temperature"],df["Pressure"],size=0.5)

show(w)
```

```python
#Styling Visual Attribute
from bokeh.plotting import figure, output_file, show

p=figure(plot_width=500,plot_height=400,tools='pan, reset')
p.title.text = "Earthquakes"
p.title.text_color = "Orange"
p.title.text_font = "times"
p.title.text_font_style = "italic"
p.yaxis.minor_tick_line_color="Yellow"
p.xaxis.axis_label = "Times"
p.yaxis.axis_label = "Value"

p.line([1,2,3,4,5],[5,6,5,5,3], line_width=2,color="red",alpha=0.5)
#OR(either of the two)
p.circle([1,2,3,4,5],[5,6,5,5,3], size = [i*2 for i in [8,12,14,15,20]])
output_file("Scatter_plotting.html")

show(p)
```

```python
import pandas
from bokeh.plotting import figure, output_file, show

df=pandas.read_csv("adbe.csv",parse_dates=["Date"])
p=figure(width=500,height=250,x_axis_type="datetime")

p.line(df["Date"],df["Close"],color="Orange",alpha=0.5)

output_file("Timeseries.html")

show(p)
```
